```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# Load the data with low_memory=False to prevent dtype warning
data = pd.read_csv('loan.csv', low_memory=False)
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
# Data Dictionary
data_dict = pd.read_excel('Data_Dictionary.xlsx')
```

```python
# Display the first few rows of the data dictionary
print(data_dict.head())
```

```python
# Data Cleaning
```

```python
# Checking for missing values
missing_values = data.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
# Handling missing values
# Dropping columns with more than 50% missing values
data = data.dropna(thresh=len(data) * 0.5, axis=1)
```

```python
# Fill remaining missing values
# Fill numerical columns with the mean value
numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns
data[numerical_cols] = data[numerical_cols].fillna(data[numerical_cols].mean())
```

```python
# Fill categorical columns with the mode value
categorical_cols = data.select_dtypes(include=['object']).columns
data[categorical_cols] = data[categorical_cols].apply(lambda x: x.fillna(x.mode()[0]))
```

```python
# Checking data types
print(data.dtypes)
```

```python
# Convert categorical variables to numerical if necessary
# Example: Convert 'term' to numerical
data['term'] = data['term'].apply(lambda x: int(x.split()[0]))
```

```python
# Replace infinite values with NaN
data.replace([float('inf'), float('-inf')], float('nan'), inplace=True)
```

```python
# Drop any remaining NaN values to ensure clean data for analysis
data.dropna(inplace=True)
```

```python
# EDA
```

```python
# Univariate Analysis
plt.figure(figsize=(10, 6))
sns.countplot(x='loan_status', data=data)
plt.title('Loan Status Distribution')
plt.show()
```

```python
# Bivariate Analysis
plt.figure(figsize=(10, 6))
sns.boxplot(x='loan_status', y='loan_amnt', data=data)
plt.title('Loan Amount vs Loan Status')
plt.show()
```

```python
# Multivariate Analysis (example: pairplot)
plt.figure(figsize=(10, 6))
sns.pairplot(data[['loan_amnt', 'int_rate', 'installment', 'loan_status']], hue='loan_status', dropna=True)
plt.show()
```

```python
# Insights and Observations
```

```python
# Calculate some basic statistics
default_rate = data['loan_status'].value_counts(normalize=True).get('Charged Off', 0)
print(f'Default Rate: {default_rate * 100:.2f}%')
```

```python
# Save the cleaned dataset for further analysis
data.to_csv('cleaned_loan_data.csv', index=False)
```
